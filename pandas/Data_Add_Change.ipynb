```python
import pandas as pd

df = pd.read_csv('cafe.csv')  # cafa.csv 만드는 방법 참고 https://kdh9797-devwiki.tistory.com/77
df.head()
```

결측치(NaN) 대입

```python
import numpy as np

df['원산지'] = np.nan
df
```

loc를 활용한 값 변경

```python
df.loc[0, '원산지'] = '콜롬비아'
df.loc[2:3, '원산지'] = '과테말라'
df.head()
```

loc를 활용한 값 추가

```python
df.loc['시즌'] = ['크리스마스라떼', 6000, 300, '한국']
df.tail()
```

loc와 딕셔너리를 활용한 값 추가

```python
df.loc[7] = {'메뉴':'달콤커피', '가격':2000, '칼로리':20}
df.tail()
```

```python
df.drop('시즌', axis=0, inplace=True)
df.to_csv('cafe2.csv', index=False)
```
