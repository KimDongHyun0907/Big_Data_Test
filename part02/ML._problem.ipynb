```python
import pandas as pd
import numpy as np

train = pd.read_csv('hr_train.csv')
test = pd.read_csv('hr_test.csv')

target = train.pop('target')

# train.select_dtypes(include='O').columns
'''
'city', 'gender', 'relevent_experience', 'enrolled_university',
       'education_level', 'major_discipline', 'experience', 'company_size',
       'company_type', 'last_new_job']
'''

# 결측치 처리
# train['gender'] = train['gender'].fillna(train['gender'].mode()[0])
# train['enrolled_university'] = train['enrolled_university'].fillna(train['enrolled_university'].mode()[0])
# train['education_level'] = train['education_level'].fillna(train['education_level'].mode()[0])
# train['major_discipline'] = train['major_discipline'].fillna(train['major_discipline'].mode()[0])
# train['experience'] = train['experience'].fillna(train['experience'].mode()[0])
# train['company_size'] = train['company_size'].fillna(train['company_size'].mode()[0])
# train['company_type'] = train['company_type'].fillna(train['company_type'].mode()[0])
# train['last_new_job'] = train['last_new_job'].fillna(train['last_new_job'].mode()[0])

train = train.fillna('X')
test = test.fillna('X')

# 인코딩
data = pd.concat([train, test])
data_oh = pd.get_dummies(data)
train = data_oh.iloc[:len(train)]
test = data_oh.iloc[:len(train)]

# 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train = scaler.fit_transform(train)
test = scaler.transform(test)

# 검증 데이터 나누기
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size = 0.2, random_state = 0)

# 학습
import lightgbm as lgb

lg = lgb.LGBMClassifier(random_state = 0, verbose = -1)
lg.fit(X_train, y_train)
pred = lg.predict_proba(X_val)

# 평가
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_val, pred[:, 1])
print(roc_auc)

# 제출
pred = lg.predict_proba(test)
submit = pd.DataFrame({'pred' : pred[:, 1]})
submit.to_csv('result.csv', index=False)
```

```python
import pandas as pd

train = pd.read_csv('creditcard_train.csv')
test = pd.read_csv('creditcard_test.csv')

target = train.pop('STATUS')

#train.isnull().sum() # OCCUPATION_TYPE

# train['OCCUPATION_TYPE'] = train['OCCUPATION_TYPE'].fillna(train['OCCUPATION_TYPE'].mode()[0])
train.fillna('X')
# train.isnull().sum()

data = pd.concat([train, test])
data_oh = pd.get_dummies(data)
train = data_oh.iloc[:len(train)]
test = data_oh.iloc[len(train):]

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
n_cols = train.select_dtypes(exclude="O").columns
train = scaler.fit_transform(train[n_cols])
test = scaler.transform(test[n_cols])

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

import lightgbm as lgb
lg = lgb.LGBMClassifier(random_state=0, verbose=-1)
lg.fit(X_train, y_train)
pred = lg.predict(X_val)

# import sklearn.metrics
# dir(sklearn.metrics)

from sklearn.metrics import f1_score
f1 = f1_score(y_val, pred)
print(f1)

pred = lg.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv('result.csv', index=False)
```

```python
import pandas as pd

train = pd.read_csv('drug_train.csv')
test = pd.read_csv('drug_test.csv')
# train.head()
target = train.pop('Drug')

data = pd.concat([train, test])
data_oh = pd.get_dummies(data)
train = data_oh.iloc[:len(train)]
test = data_oh.iloc[len(train):]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
n_cols = train.select_dtypes(exclude="O").columns
train = scaler.fit_transform(train[n_cols])
test = scaler.transform(test[n_cols])

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size = 0.2, random_state = 0)

import lightgbm as lgb
lg = lgb.LGBMClassifier(random_state = 0, verbose = -1)
lg.fit(X_train, y_train)
pred = lg.predict(X_val)

# import sklearn.metrics
# dir(sklearn.metrics)
from sklearn.metrics import f1_score
f1 = f1_score(y_val, pred, average = 'macro')
print(f1)

pred = lg.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv('result.csv', index=False)
```
